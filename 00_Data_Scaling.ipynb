# Import Required Libraries
Import the necessary libraries, including NumPy, Matplotlib, and Scikit-learn.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# Create a Dummy Dataset
Create a dummy dataset using NumPy's random number generator. What is the shape of the dataset? What does the data look like?

```python
# Set the seed for the random number generator for reproducibility
np.random.seed(0)

# Generate a random dataset with 100 samples and 1 feature
X = np.random.rand(100, 1) * 10
y = np.random.rand(100, 1) * 10

# Print the shape of the dataset
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)

# Print the first 5 samples of the dataset
print("First 5 samples of X:", X[:5])
print("First 5 samples of y:", y[:5])
```

# Plot the original data


```python
plt.scatter(X, y, color='blue')
plt.title('Original Data')
plt.tight_layout()
plt.show()
```

# Apply StandardScaler
Apply the StandardScaler to the dataset. What does this do to the data? How does it change the distribution of the data? Try changing the parameters of the StandardScaler and observe the effects.

```python
# Instantiate the StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the X data and transform it
X_scaled = scaler.fit_transform(X)

# Fit the StandardScaler to the y data and transform it
y_scaled = scaler.fit_transform(y)

# Print the first 5 samples of the scaled data
print("First 5 samples of X_scaled:", X_scaled[:5])
print("First 5 samples of y_scaled:", y_scaled[:5])
```

# Plot the original and scaled data


```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].scatter(X, y, color='blue')
axes[0].set_title('Original Data')
axes[1].scatter(X_scaled, y_scaled, color='red')
axes[1].set_title('StandardScaler Data')
for ax in axes:
    ax.set_xlabel('X')
    ax.set_ylabel('y')
plt.tight_layout()
plt.show()

# Question: What does the StandardScaler do to the data?
# Answer: The StandardScaler standardizes the features by removing the mean and scaling to unit variance.

# Question: How does the StandardScaler change the distribution of the data?
# Answer: The StandardScaler does not change the shape of the distribution of the data. It only changes the scale of the data.

# Question: Try changing the parameters of the StandardScaler and observe the effects.
# Task: Change the parameters of the StandardScaler and observe the effects on the scaled data.
```

# Apply MinMaxScaler
Apply the MinMaxScaler to the dataset. How does this compare to the StandardScaler? What are the differences? Try changing the parameters of the MinMaxScaler and observe the effects.

```python
# Instantiate the MinMaxScaler
minmax_scaler = MinMaxScaler()

# Fit the MinMaxScaler to the X data and transform it
X_minmax_scaled = minmax_scaler.fit_transform(X)

# Fit the MinMaxScaler to the y data and transform it
y_minmax_scaled = minmax_scaler.fit_transform(y)

# Print the first 5 samples of the min-max scaled data
print("First 5 samples of X_minmax_scaled:", X_minmax_scaled[:5])
print("First 5 samples of y_minmax_scaled:", y_minmax_scaled[:5])
```

# Plot the Data
Plot the original, StandardScaler, and MinMaxScaler data. What differences do you observe between the three plots? How do the scales compare?

```python
# Plot the original, StandardScaler and MinMaxScaler data
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].scatter(X, y, color='blue')
axes[0].set_title('Original Data')
axes[1].scatter(X_scaled, y_scaled, color='red')
axes[1].set_title('StandardScaler Data')
axes[2].scatter(X_minmax_scaled, y_minmax_scaled, color='green')
axes[2].set_title('MinMaxScaler Data')
for ax in axes:
    ax.set_xlabel('X')
    ax.set_ylabel('y')
plt.tight_layout()
plt.show()
```
