### 1.  Import Required Libraries
Import the necessary libraries, including NumPy and Matplotlib.

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
```

### 2.  Generate Random Data
Generate random data for two classes using the NumPy library.

```python
# Generate random data for class 1
class1_x = np.random.uniform(0, 5, 5)
class1_y = np.random.uniform(0, 5, 5)

# Generate random data for class 2
class2_x = np.random.uniform(4, 9, 5)
class2_y = np.random.uniform(4, 9, 5)

# Combine the data points and labels
data = np.concatenate((np.array([class1_x, class1_y]).T, np.array([class2_x, class2_y]).T))
labels = np.concatenate((np.ones(len(class1_x)), -np.ones(len(class2_x))))
```

### 3.  Visualize the Dataset
Use Matplotlib to visualize the dataset.

```python
# Set the limits for the x and y axes
plt.xlim(0, 9)
plt.ylim(0, 9)

# Plot the data points for class 1
plt.scatter(class1_x, class1_y, color='red', label='Class 1')

# Plot the data points for class 2
plt.scatter(class2_x, class2_y, color='blue', label='Class 2')

# Set the labels for the x and y axes
plt.xlabel('X')
plt.ylabel('Y')

# Set the title for the plot
plt.title('Linearly Separable Dataset')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

### 4.  Combine Data Points and Labels
Combine the data points and labels for the two classes.

```python
# Combine the data points and labels
data = np.concatenate((np.array([class1_x, class1_y]).T, np.array([class2_x, class2_y]).T))
labels = np.concatenate((np.ones(len(class1_x)), -np.ones(len(class2_x))))

# Print the combined data and labels for verification
print("Combined Data Points:")
print(data)
print("\nCombined Labels:")
print(labels)
```

### 5.  Initialize Weights and Bias
Initialize the weights and bias for the Perceptron learning algorithm.

```python
# Initialize the weights and bias
weights = np.zeros(2)
bias = 0

# Print the initial weights and bias
print("Initial Weights:", weights)
print("Initial Bias:", bias)
```

### 6.  Define Activation Function
Define the activation function for the Perceptron learning algorithm.

```python
# Define the activation function
def activation_function(x):
    return np.where(x >= 0, 1, -1)

# Test the activation function with some inputs
print("Activation Function Output for -10:", activation_function(-10))
print("Activation Function Output for 0:", activation_function(0))
print("Activation Function Output for 10:", activation_function(10))
```

### 7.  Set Learning Rate and Number of Iterations
Set the learning rate and number of iterations for the Perceptron learning algorithm.

```python
# Set the learning rate and number of iterations
learning_rate = 1
num_iterations = 100

# Print the learning rate and number of iterations
print("Learning Rate:", learning_rate)
print("Number of Iterations:", num_iterations)
```

### 8.  Perceptron Learning Algorithm
Implement the Perceptron learning algorithm to classify the data points.

### Visualize Decision Boundary
Visualize the decision boundary after each iteration of the Perceptron learning algorithm.

```python
# Perceptron Learning Algorithm
for iteration in range(num_iterations):
    misclassified_points = 0
    for i in range(len(data)):
        x = data[i]
        y = labels[i]
        prediction = activation_function(np.dot(weights, x) + bias)
        if prediction != y:
            misclassified_points += 1
            weights += learning_rate * y * x
            bias += learning_rate * y
    print(f"Iteration: {iteration+1}")
    print(f"Number of misclassified points: {misclassified_points}")
    print(f"Weights: {weights}")
    print(f"Bias: {bias}")
    print(f"Decision boundary: {weights[0]}x + {weights[1]}y + {bias} = 0")
    print(f"Normal vector to the decision boundary: [{weights[0]}, {weights[1]}]")
    print("------------------------------")
    
    plt.xlim(0, 9)
    plt.ylim(0, 9)
    # Plot the dataset
    plt.scatter(class1_x, class1_y, color='red', label='Class 1')
    plt.scatter(class2_x, class2_y, color='blue', label='Class 2')
    
    # Plot the decision boundary
    x_vals = np.linspace(0, 9, 100)
    y_vals = -(weights[0] * x_vals + bias) / weights[1]
    plt.plot(x_vals, y_vals, color='green', label='Decision Boundary')
    
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Linearly Separable Dataset')
    plt.legend()
    plt.show()
    
    if misclassified_points == 0:
        break
```
