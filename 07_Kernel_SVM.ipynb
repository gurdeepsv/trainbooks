### 1.  Import Required Libraries
Import the necessary libraries, including NumPy, sklearn.svm, matplotlib.pyplot, and sklearn.datasets.

```python
# Importing the required libraries
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

### 2.  Create Linearly Inseparable Points
Use the make_blobs function from sklearn.datasets to create a set of linearly inseparable points.
#### TODO: Try generating different Inseparable Points

```python
# Create linearly inseparable points
X, y = make_blobs(n_samples=18, centers=[(2, 2), (4, 4)])

# Plot the points
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
plt.show()
```

### 3.  Fit the SVM Model
Create an instance of the SVC class with a polynomial kernel, and fit the model to the data.
#### TODO: Try different degrees for poly kernel
#### TODO: Try different C parameter for poly kernel
#### TODO: Try different kernels

```python
# Create an instance of the SVC class with a polynomial kernel
svm = SVC(kernel='poly', degree=4, C=10**15)

# Fit the model to the data
svm.fit(X, y)
```

### 4.  Visualize the Dataset Again
Visualize the dataset again after fitting the model.

```python
# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# Plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

# Plot support vectors
ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k')

# Show the plot
plt.show()
```
