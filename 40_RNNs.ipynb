### 1.  Section 1: Usage of Embedding Layer
This section covers the usage of the Embedding Layer for Text Data. It includes data preparation, creation of ANN using an embedded layer, model compilation and fitting, inline prediction and evaluation, computing the accuracy, and out of sample prediction.

```python
# Import necessary libraries
import numpy as np
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Dense, Flatten
from keras.metrics import BinaryAccuracy
```

### 2.  Define the reviews


```python
reviews =['never coming back!',
'horrible service',
'rude waitress',
'cold food',
'horrible food!',
'awesome',
'awesome services!',
'rocks',
'poor work',
'couldn\'t have done better'
]
```

```python
# Initialize the tokenizer
t=Tokenizer()
t.fit_on_texts(reviews) # tokenizing the words
```

```python
# Encode the documents
encoded_docs=t.texts_to_sequences(reviews)
```

```python

# Pad the reviews
max_length=4
padded_reviews=pad_sequences(encoded_docs,maxlen=max_length,padding='post')
```

```python

# Define the labels
y=np.array([0,0,0,0,0,1,1,1,0,1])
```


### Create the model


```python
model=Sequential()
voc_size=19 # one added with the number of unique terms
emb_layer=Embedding(input_dim=voc_size,output_dim=10,input_length=max_length)
model.add(emb_layer)
model.add(Flatten())
model.add(Dense(16,activation='relu')) # addding hidden layer
model.add(Dense(1,activation='sigmoid'))
```


### Compile the model


```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
```


### Fit the model


```python
model.fit(padded_reviews,y,epochs=50)
```

```python

# Inline prediction and evaluation
y_pr=np.round(model.predict(padded_reviews))
```

```python

# Compute the accuracy
bac=BinaryAccuracy()
acc=bac(y,model.predict(padded_reviews)).numpy()
```

```python

# Out of sample prediction
test=np.array([[10,1,0,0]]) # Predict the class of the sentence is "cold horrible"
print(np.round(model.predict(test)[0,0]))
```

### 6.  Section 2: Word Prediction using RNN
This section covers the application of RNN in word prediction. It includes preprocessing of the data, construction of RNN, model compilation, fitting and prediction, and out of sample prediction.

```python
# Import necessary libraries
from keras.layers import SimpleRNN
from keras.utils import to_categorical
```

```python

# Define the data
data="Oman, a jewel of the Arabian Peninsula, stuns visitors with its dramatic desert landscapes, turquoise waters, and ancient forts guarding historic treasures. From the rugged peaks of the Jebel Akhdar mountains to the vibrant coral reefs teeming with life, Oman offers a captivating blend of natural beauty and rich cultural heritage."
```

```python

# Initialize the tokenizer
t=Tokenizer()
t.fit_on_texts([data])
```

```python

# Get the word index
wo_indx=t.word_index
```

```python

# Numeric encoding of the data
encoded_data=t.texts_to_sequences([data])[0] 
```

```python

# Create feature and target values for predicting a word when a single word is given
n=len(encoded_data) # to know the number of items in the sequence
dt_seq=list()
for i in range(1,n):
    seq=encoded_data[i-1:i+1]
    dt_seq.append(seq)
```

```python

# Convert the sequence to a numpy array
seq=np.array(dt_seq)
```

```python

# Split the sequence into features and target
X,y=seq[:,0],seq[:,1]

# Convert the target to categorical
voc_size=len(t.word_index)+1
y_cat=to_categorical(y,num_classes=voc_size)
```

```python

# Create the RNN model
model=Sequential()
model.add(Embedding(voc_size,10,input_length=1)) # meant for predicting one word given the previous one
model.add(SimpleRNN(50))
model.add(Dense(voc_size,'softmax'))
```

```python

# Compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python

# Fit the model
model.fit(X,y_cat,epochs=500)
```

```python

# Predict the words
pred=np.argmax(model.predict(X),axis=1)
```

```python

# Get the words
word=list(wo_indx.keys()) # collect the keys and store it as a list
```

```python

# Create the predicted words
pr_words=''
for i in range (1,len(pred)+1):
    pr_words=pr_words+' ' + word[pred[i-1]-1]
```
